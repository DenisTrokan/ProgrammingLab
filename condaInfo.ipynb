# 🐍 Basic Conda Commands Cheat Sheet

---

### ✅ 1. Check Conda is installed
```
conda --version
```

---

### 📦 2. Create a new environment
```
conda create -n myenv python=3.11
```
- `myenv`: name of the environment  
- `python=3.11`: optional version

---

### ▶️ 3. Activate an environment
```
conda activate myenv
```

---

### ⏹️ 4. Deactivate the environment
```
conda deactivate
```

---

### 🔍 5. List all environments
```
conda env list
```

---

### 📜 6. Install a package
```
conda install numpy
```

Install multiple packages:
```
conda install numpy pandas matplotlib
```

---

### 🔄 7. Update a package
```
conda update numpy
```

---

### ❌ 8. Remove a package
```
conda remove numpy
```

---

### 🧼 9. Delete an environment
```
conda remove -n myenv --all
```

---

### 📂 10. Export environment to `.yml`
```
conda env export > environment.yml
```

---

### 📥 11. Create environment from `.yml`
```
conda env create -f environment.yml
```

---

### 💡 Bonus Tips
```
conda list     # View installed packages
conda info     # Show Conda system info
```
